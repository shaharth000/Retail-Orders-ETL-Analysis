```python
#extract file from the zipfile
import zipfile

zip_ref=zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#read the file using pandas

import pandas as pd

df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head()
```

```python
#Handling null values
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
#rename columns, making them in lower case and replacing space with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#derive new columns, discount, sell price and profit

df['discount_price'] = df['list_price']*df['discount_percent']/100
df['sell_price'] = df['list_price'] - df['discount_price']
df['profit'] = df['sell_price'] - df['cost_price']
```

```python
df.head()
```

```python
df.dtypes
```

```python
#convert order date from object to datetime

df['order_date']= pd.to_datetime(df['order_date'],format = '%Y-%m-%d')
```

```python
df.dtypes
```

```python
#drop unwanted columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
